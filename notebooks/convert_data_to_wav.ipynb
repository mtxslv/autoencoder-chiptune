```python
from pathlib import Path
import subprocess
```

```python
# Data folders
__name__ = 'convert_data_to_wav.ipynb'
raw_data_filepath = Path(__name__).resolve().parents[1] / 'data' / 'raw'
assert raw_data_filepath.exists()
mp3_files_folder = raw_data_filepath / 'mp3'
wav_files_folder = raw_data_filepath / 'wav'
assert mp3_files_folder.exists()
assert wav_files_folder.exists()
```

```python
# Get mp3 files
mp3_files = [file for file in mp3_files_folder.glob("*.mp3")]
assert len(mp3_files) == 50
for mp3_file in mp3_files:
    assert mp3_file.is_file()
    assert mp3_file.exists()
```

```python
# Convert mp3 to wav
for mp3_file in mp3_files:
    file_name = mp3_file.name.split('.mp3')[0] + '.wav'
    output_path = wav_files_folder / file_name
    if not output_path.exists():
        subprocess.call(['ffmpeg', 
                         '-i', 
                         str(mp3_file),
                         str(output_path)])
```

```python
# Check if wav files are ok
wav_files = [file for file in wav_files_folder.glob("*.wav")]
assert len(wav_files) == 50
for wav_file in wav_files:
    assert wav_file.is_file()
    assert wav_file.exists()
```
